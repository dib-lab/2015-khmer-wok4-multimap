```python
import numpy
```

```python
def load_counts(filename, limit=2000):
    counts = {}
    for n, line in enumerate(open(filename)):
        if limit and n > limit:
            break
        line = line.strip().split(' ')
        name = line[0]
        ctl = [ float(c) for c in line[1:] ]
        counts[name] = ctl
    return counts


def build_array(mapping_filename, meds_filename):
    mapped = load_counts(mapping_filename, limit=None)
    meds = load_counts(meds_filename)
    
    xx = []
    for k in meds:
        row = [mapped.get(k, [0])[0]]
        row.extend(meds[k])
        xx.append(row)
        
    return numpy.array(xx)
```

```python
rseq = build_array('rseq-mapped.counts', 'rseq.labelcount')
```

```python
loglog(rseq[:,1], rseq[:,0], 'g+', label='count')

title('camille\'s awesome mapping')
xlabel('khmer graphalign')
ylabel('bowtie2')
legend(loc='upper left')
savefig('fig1.png')
```
